### 영상처리

- 입력된 영상을 어떤 목적을 위해 수학적 연산을 이용하여 처리하는 기술
    - 어두운 영상의 각 화소에 밝기 값 50을 더하는 연산을 통해 밝은 영상을 만드는 것.
    -  처리 결과에 따라 두 가지로 나뉜다.
        > 저수준 영상처리; 처리하고 난 결과가 영상  
        고수준 영상처리; 처리하고 난 결과가 영상이 아니라 특성

### CV & CG

# Computer Vision

- 현실의 영상이나 이미지를 취득해서 정보를 얻는 기술
- input으로 이미지나 비디오를 받아 output으로 그 이미지나 비디오로부터 얻을 수 있는 정보를 얻는 기술
    - 2D 이미지를 해석하거나 3D 장면을 해석하는 것.

Computer VIsion
> Recognition  
Detection  
Segmentation  
Tracking


=> 사람의 시각 시스템을 computer로 모델링하거나 구조화하는것. 즉, 사람처럼 이미지를 보고 그 이미지에 대한 정보를 얻어내는 것.

# Computer Graphics

- 실제 사진이나 영상을 사용하지 않았음에도 현실과 같은 가상의 영상을 만들어 내는 것.

- input으로 원하는 정보를 입력하면(원이 어느정도의 크기로 어느 지점에있고, 그 안에 검은색... 등등) output으로 가상의 영상을 얻을 수 있는 기술.

Computer graphics
> Retargeting  
Image Manupulation  
Rendering

=> 기하학적 객체의 이미지 생성, 조작 및 저장을 위한 기술
