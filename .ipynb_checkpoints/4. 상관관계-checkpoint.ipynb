# 상관관계

- 하나의 변수가 아닌, 두 변수 사이의 상호관계를 분석하기 위한 방법.  
    - 하나의 변수에 대해서 분석할 때는 Histogram

- 중간고사와 기말고사 성적이 있다.
    - 두 변수간의 관계를 파악하기위해서는 '결합분포'를 봐야한다.
    - 산포도(scatter plot)을 이용한다.
    


# 상관계수

- 산포도에서 점들의 부포가 타원형을 띈다고한다.

- 1에 가까운 상관계수는 산포도 점들이 하나의 선 주위에 몰려 있다. (A)

- 0에 가까운 상관계수는 점들이 처져있다. (B)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'C://users/anjae/Documents/[0] Dataset/medical_insurance/insurance.csv'
DF = pd.read_csv(path)
DF.head()
```

# 상관관계1 - 나이에 따른 charges

```python
x = DF['age']
y = DF['charges']
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(x, y,s =1)

ax.set_xlabel('age')
ax.set_ylabel('charges')
```

```python
x_bar = np.mean(x)
y_bar = np.mean(y)

std_x = np.sqrt( np.sum((x-x_bar)**2) / len(x)-1 )
std_y = np.sqrt( np.sum((y-y_bar)**2) / len(y)-1 )

cov_xy = np.sum( (x-x_bar) * (y-y_bar) ) / (len(x) -1)
r = cov_xy / (std_x*std_y)

r
```

```python
np.corrcoef(x, y)
```

# 여러가지 상관계수

- pearson correlation coefficient: 두 변수가 모두 연속형 자료일 때 유용


- spearman correlation coefficient: 두 연속형 변수의 분포가 심각하게 정규분포를 벗어나거나
                                    - 두 변수가 순위 척도(ordinal scale)일 때
                                    
- kendall correlatino coefficient: 순위 척도 자료 or 순위 척도 변환한 자료를 이용해 두 변수간 상관계수를 알아 볼 때

# 상관관계 ellipse

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
```

#### np.cov(x,y) 의 return에 대해서

- np.cov는 return값이 2*2이다.

cov(a, a)    cov(a, b)  
cov(a, b)    cov(b, b)  

- 1 * 1 은 변수 a의 분산

- (1 * 2) (2 * 1) 은 변수 a와 b의 공분산

- (2 * 2) 는 변수 b의 분산

```python
def norm(data):
    return (data-np.mean(data)) / np.std(data)
```

```python
def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs):
    
    if x.size != y.size:
        raise ValueError("x and y must ne the same size")
        
    cov = np.cov(x, y)
    pearson = cov[0, 1]/np.sqrt(cov[0, 0]*cov[1,1])  # 연속형 변수라 가정 후 pearson 상관 계수 구하기
    
    
    ell_radius_x = np.sqrt( 1+ pearson ) # pearson 계수에 1을 더해준 것의 제곱근
    ell_radius_y = np.sqrt( 1- pearson ) # pearson 계수에 -1을 더해준 것의 제곱근
    ellipsis = Ellipse((0,0),
                       width = ell_radius_x * 2,    # total length of horizontal axis // 길이
                       height = ell_radius_y * 2,   # total length of vertical axis   // 길이
                      facecolor=facecolor,
                      **kwargs)
    
    
    scale_x = np.sqrt(cov[0, 0]) * n_std           # x의 분산 에 표준편차 곱한 것
    mean_x = np.mean(x)
    
    scale_y = np.sqrt(cov[1,1]) * n_std
    mean_y = np.mean(y)
    
    transf = transforms.Affine2D().rotate_deg(45).scale(scale_x, scale_y).translate(mean_x, mean_y)
    
    ellipsis.set_transform(transf + ax.transData)
    return ax.add_patch(ellipsis)
```
